# Bell State

Create and simulate a two-qubit Bell state, the basic entangled state.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
bell = QuantumCircuit(2, 2)
bell.h(0)
bell.cx(0, 1)
bell.measure([0, 1], [0, 1])

bell.draw("mpl")
```

```python
simulator = AerSimulator()
compiled = transpile(bell, simulator)
result = simulator.run(compiled, shots=1000).result()
counts = result.get_counts()

counts
```

```python
plot_histogram(counts)
```

# Notes
Measures either 00 or 11, as expected for an entangled Bell state. Transpiled simulators lead to better results (i.e. closer 50/50 ratio) than non-transpiled.
